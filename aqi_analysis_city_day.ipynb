```python
df = pd.read_csv('./aqi_datasets/city_day.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Datetime'])
df.head()
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df.head()
```

```python
df['City'].unique()
```

```python
df['AQI'].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_cities = df.groupby('City')['AQI'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.title('Top 10 Cities by Average AQI')
plt.xlabel('Average AQI')
plt.ylabel('City')
plt.show()
```

```python
yearly_trend = df.groupby('Year')['AQI'].mean()
print(yearly_trend)
sns.lineplot(x=yearly_trend.index, y=yearly_trend.values, marker='o')
plt.title('Yearly Trend of AQI')
plt.show()
```

```python
sns.boxplot(x='Month',y='AQI',data=df)
plt.title('Air Quality Index by Month')
plt.show()
```

```python
city_year = df.groupby(['City',"Year"])['AQI'].mean().unstack()
plt.figure(figsize=(14,10))
sns.heatmap(city_year,cmap='Reds',linewidths=0.5)
plt.title('Average AQI by City and Year')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='AQI_Bucket',data=df,order=df['AQI_Bucket'].value_counts().index,palette='coolwarm')
plt.title('Distribution of AQI Bucket')
plt.xticks(rotation=45)
plt.show()
```

```python
pollutent_means = df[['PM2.5', 'PM10','NO2','SO2','O3','CO']].mean()
print(pollutent_means)
pollutent_means.sort_values(ascending=False).plot(kind='bar',color='orange')
plt.title('Mean of Pollutants')
plt.xlabel('Pollutants')
plt.ylabel('Mean')
plt.show()
```

```python
cities = df['City'].unique()
window_size = 30

for city in cities:
    city_df = df[df['City'] == city].set_index('Date').sort_index()
    city_df['AQI_Rolling'] = city_df['AQI'].rolling(window=window_size).mean()

    plt.figure(figsize=(12,6))
    plt.plot(city_df.index, city_df['AQI'], label='Daily AQI', alpha=0.5)
    plt.plot(city_df.index, city_df['AQI_Rolling'], label=f'{window_size}-Day Rolling Avg',color='red',linewidth=2)
    plt.title(f'{city} AQI with {window_size}-Day Rolling Average')
    plt.xlabel('Date')
    plt.ylabel('AQI')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
